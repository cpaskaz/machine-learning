### Selecting Models

For a supervised regression use case, there are numerous algorithms you can use, each with its own strengths and suitability for different kinds of data and problem complexities. Here's a list of commonly used Tree and Ensemble based regression algorithms along with a brief description of each:

1. **Decision Trees**
  - Type: Tree-based model
  - Use Case: Binary or multiclass classification, or continous targets
  - Pros: Easy to interpret, can handle non-linear data
  - Cons: Prone to overfitting
2. **Random Forest**
  - Type: Ensemble of Decision Trees

Performance requirements: If interpretability is key, decision trees or logistic regression might be preferred. If the highest accuracy is the goal, more complex models like gradient boosting machines or neural networks could be better choices.

Computational resources: Some algorithms, especially neural networks and ensemble methods, require significant computational power and training time.

It's often beneficial to experiment with multiple algorithms and perform cross-validation to determine which algorithm performs best for your specific use case.

### Selecting Metrics

Evaluating regression models involves assessing how well the model's predictions approximate the actual data points. Several metrics are used to quantify this, each highlighting different aspects of model performance. Here's a summary of key metrics commonly used in regression model evaluation:

1. **Mean Absolute Error (MAE)**
   - ***Type:*** Error metric
   - ***Use Case:*** Measures the average magnitude of errors in a set of predictions, without considering their direction. It’s a linear score, meaning all individual differences are weighted equally in the average.
   - ***Pros:*** Easy to understand; treats all errors equally.
   - ***Cons:*** May not highlight significant errors sufficiently because it doesn't square the errors.
2. **Mean Squared Error (MSE)**
   - ***Type:*** Error metric
   - ***Use Case:*** Measures the average of the squares of the errors. It’s more sensitive to outliers than MAE because it squares the differences.
   - ***Pros:*** Emphasizes larger errors more than smaller ones, making it useful when large errors are particularly undesirable.
   - ***Cons:*** Because it squares the errors, it can dramatically increase the impact of outliers on the evaluation.
3. **Root Mean Squared Error (RMSE)**
   - ***Type:*** Error metric
   - ***Use Case:** It’s the square root of the MSE and provides a measure of the magnitude of the error in the same units as the response variable.
   - ***Pros:*** More interpretable in the context of the target variable than MSE; emphasizes larger errors.
   - ***Cons:*** Still sensitive to outliers, like MSE.
4. **R-squared (Coefficient of Determination)**
   - ***Type:*** Goodness-of-fit metric
   - ***Use Case:*** Represents the proportion of the variance in the dependent variable that is predictable from the independent variables. It’s a measure of how well the regression predictions approximate the real data points.
   - ***Pros:*** Easy to interpret; values range from 0 to 1, where 1 indicates perfect prediction.
   - ***Cons:*** Can be misleading if the model is overfitted; does not necessarily mean the model is good if R-squared is high without considering other metrics.
5. **Adjusted R-squared**
   - ***Type:*** Goodness-of-fit metric
   - ***Use Case:*** Adjusts the R-squared value based on the number of predictors in the model, providing a metric that doesn’t improve by merely adding more predictors.
   - ***Pros:*** More accurate reflection of the model’s explanatory power when comparing models with different numbers of predictors.
   - ***Cons:*** Still does not indicate whether a regression model is adequate; needs to be used alongside other metrics.
6. **Mean Absolute Percentage Error (MAPE)**
   - ***Type:*** Error metric
   - ***Use Case:*** Measures the average magnitude of the errors in percentage terms. It’s useful when you want to know the error as a percentage of the actual values.
   - ***Pros:*** Provides error magnitude in understandable terms (percentages); good for comparing accuracy across different datasets or models.
   - ***Cons:*** Can be infinite or undefined if there are actual values equal to zero; heavily biased towards predictions that are systematically lower than actual values.
7. **Mean Squared Logarithmic Error (MSLE)**
   - ***Type:*** Error metric
   - ***Use Case:*** Useful when you do not want to penalize large differences when both the predicted and actual values are large numbers.
   - ***Pros:*** Can be useful in situations where the target has exponential growth, e.g., population counts, or when you care more about percentage errors.
   - ***Cons:*** Not as intuitive as other metrics; unsuitable for datasets with negative values.

When selecting metrics for evaluating regression models, consider the specific characteristics of your data, the impact of outliers, and whether over- or under-estimations are penalized differently. Often, using a combination of these metrics provides a more holistic view of model performance.

### Decision Tree - Tuned Parameter Hyper-Tuning

```python
with parallel_backend(backend='multiprocessing', n_jobs=-1):

    # Choose the type of classifier
    dtree_estimator = DecisionTreeRegressor(random_state = 1)
    # Grid of parameters to choose from
    parameters = {
        "max_depth": [1,3,5,7,9,11],
        "min_samples_leaf": [1,2,3,4,5,6,7,8,19,10],
        "max_leaf_nodes": [None, 10, 20, 30, 40],
    }

    # Type of scoring used to compare parameter combinations
    #scorer = metrics.make_scorer('neg_mean_squared_error')
    
    # Run the grid search
    gridCV = GridSearchCV(dtree_estimator, parameters, scoring = 'neg_mean_squared_error', cv = 5)

    # Fitting the grid search on the train data
    gridCV = gridCV.fit(x_train, y_train)

    # Set the classifier to the best combination of parameters
    dtree_estimator = gridCV.best_estimator_
```
